```python
import pandas as pd
```

```python
import numpy as np
```

```python
import re
```

```python
train=pd.read_csv('washed_train.csv',sep=',')
```

```python
train['A20'].apply(lambda x:re.findall('\d+',x))
```

```python
def get_duration(vt):
    ct=str(vt)
    if len(re.findall('.+-.+',ct))>0:
        serial=re.findall('\d+',ct)
        for x in range(0,len(serial)):
            serial[x]=int(serial[x])
        if len(serial)==3:
            serial.append(0)
        if serial[0]>serial[2]:
            serial[2]=serial[2]+24
        return (serial[2]-serial[0])*60+serial[3]-serial[1]
    else:
        return vt
```

```python
pde=train.applymap(get_duration)
```

```python
pde.head(12)
```

```python
pde.to_csv('washed2_train.csv',sep=',',index=0)
```
